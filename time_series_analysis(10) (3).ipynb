```python
# now if you want to know the average stock of every month of any year

# you need to use resample function for that

# but before that you need to select the year(using slicing), then choose a column(Close in our case)on which you want to do resample

# then find the mean

df['2010-01':'2010-12'].Close.resample('M').mean()
```

```python

df['2010-01':'2010-12'].Close.resample('M')
```

```python
df['2014-01':'2015-12'].Close.resample('M').sum()
```

```python
df['2014-04-01':'2015-03-31'].Close.resample('M').sum()
```

```python
df['2014-04-01':'2014-01-01'].Close.resample('M').sum()
```

```python
# you can also plot the chart of any month stock average

%matplotlib inline
df1 = df['2011-01':'2011-12'].Close.resample('M').mean().plot()
df1
```

```python
df["2011-01":"2011-12"]
```

```python
# you can also watch the weekely frequency by just giving the value "W" to resample()

df["2012-01":"2012-05"].Close.resample("W").mean()
```

```python
%matplotlib inline
df["2012-01":"2012-05"].Close.resample("W").mean().plot()
```

```python
# suppose you want to know the semi-month start frequency (1st and 15th) of any year

df["2013-01":"2013-12"].Close.resample("SMS").mean()
```

```python
# supoose i want to draw bar plot for the above frequency then use plot(kind="bar")

%matplotlib inline
df["2013-01":"2013-12"].Close.resample("SMS").mean().plot(kind="bar")
```

```python
# using resample() you can able to see you data on different ranges

# se the difference of with resample or without resample, average stock 2011

# without resample

%matplotlib inline
df["2011-01"].Close.plot()
```

```python
# with using resample

%matplotlib inline
df["2011-01"].Close.resample("W").plot()
```
